```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
stud=pd.read_csv("C:\\FCD\\PythonFundamental_Femsa\\clean_students_complete.csv")
```

```python
stud.set_index("Student ID",inplace=True)
```

```python
stud=stud.drop(["Unnamed: 0"],axis=1)
```

```python
menor_pt=stud[stud["reading_score"]==stud["reading_score"].min()] #menor pontuação em reading
menor_pt
```

```python
menor_pt_gen_school=menor_pt[["gender","school_name"]] #genero e escola dos alunos com menor pt
```

```python
dum_gen = pd.get_dummies(menor_pt_gen_school[["gender"]])
```

```python
plotgen=pd.concat([menor_pt_gen_school,dum_gen],axis=1)
```

```python
plotgroup=plotgen.groupby(["school_name"]).sum()
```

```python
plotgroup.plot.bar(figsize=(10,4))
```
